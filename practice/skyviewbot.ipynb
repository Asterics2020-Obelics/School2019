### Make it interactive

```python
from skyviewbot import call_skyview, coords_from_name, plot_fits
```

```python
call_skyview("DSS", coords_from_name("M33"), 1, "J2000", "M33.fits")
```

```python
from ipywidgets import interact, interactive
```

```python
def plot_wrapper(colormap="viridis"):
    plot_fits("M33.fits", "M33", colormap, False, None)
```

```python
interactive_plot = interactive(plot_wrapper, colormap=['viridis', 'plasma', 'inferno', 'magma'])
output = interactive_plot.children[-1]
output.layout.height = '500px'
interactive_plot
```
